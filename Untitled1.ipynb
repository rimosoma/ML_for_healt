```python

"""
@author: Simone Romano

We want to regress again Total UPDRS as in laboratory # 1, and we will only use LLS method. The
difference is that we want to use only K nearest neighbours to train the model. We will use the
validation dataset to find the ‚Äúoptimum‚Äù value of K.



Regress Total UPDRS from the other features in file "parkinsons_updrs_av.csv"
This lab implements a K-Nearest-Neighbor LLS (local regression) model
and compares it to a standard LLS model.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.precision', 3)



#%% Read the input csv file
plt.close('all')
X=pd.read_csv("parkinsons_updrs_av.csv")
features=list(X.columns)
subj=pd.unique(X['subject#'])

# --- CONFIGURATION ---
# flag to enable/disable data shuffling before train/test split
SHUFFLE_DATA = True

ALL_INFO_FLAG = True

# Small constant for Ridge Regression (matrix inversion stability)
epsilon = 10e-8

Np,Nc=X.shape
```

```python
#STARTIBG ANALYSIS


if ALL_INFO_FLAG is True:
    print("---------------------------------------")
    print("---------------------------------------")
    print('‚úÖALL_INFO_FLAG: ON; ‚¨áÔ∏è‚¨áÔ∏è‚¨áÔ∏è‚¨áÔ∏è‚¨áÔ∏è‚¨áÔ∏è‚¨áÔ∏è')
    print(" ")
    print("-------------------------------------------------------------------------")
    print("-------------------------------------------------------------------------")
    print("-------------------üß†‚Äã BASIC DATABASE INFO üß†‚Äã-----------------------------")
    print("-------------------------------------------------------------------------")
    print("-------------------------------------------------------------------------")
    # print basic dataset info
    print("The original dataset shape  is ",X.shape)
    print("The number of distinct patients in the dataset is ",len(subj))
    print("the original dataset features are ",len(features))
    #print(features)

    #%% Have a look at the dataset
    print("---------------------------------------")
    print("---------------------------------------")
    print("HERE THE ORIGINAL FEATURES WITH THEIR DESCRIPTIONS:‚¨áÔ∏è‚Äã")
    print(X.describe().T)
    # show dtypes and non-null counts (helps detect missing data)
    print("---------------------------------------")
    print("---------------------------------------")
    print("HERE DEFAULT INFORMATION ABOUT THE ORIGINAL X:‚¨áÔ∏è‚Äã")
    print(X.info())


#FINISH ANALYSIS
```

```python
#START PREPARING DATASET

print("-------------------------------------------------------------------------")
print("-------------------------------------------------------------------------")
print("-------------------üß†‚Äã PREPARATION PHASE üß†‚Äã-----------------------------")
print("Split the data in train/val/test and normalize them using train set data")
print("-------------------------------------------------------------------------")

if ALL_INFO_FLAG is False:
    print("---------------------------------------")
    print("---------------------------------------")
    print('‚ùóALL_INFO_FLAG: OFF')


#%% Shuffle the data
if SHUFFLE_DATA:
    Xsh = X.sample(frac=1, replace=False, random_state=310436, axis=0, ignore_index=True)
    print('‚úÖ‚ÄãSHUFFLING FLAG: ON (random_state=310866)')

else:
    Xsh = X.copy()
    print('‚ùóSHUFFLING FLAG: OFF (data kept in original order)')


#%% Define the lenghts of the sets and define the train set
# Split data as per instructions (Page 5, point 3): 40% Train, 20% Validation, 40% Test
Ntr=int(Np*0.4)
Nval=int(Np*0.2)
Nte=Np-Ntr-Nval
print("---------------------------------------")
print("---------------------------------------")
print(f"Data split: N_Train={Ntr}, N_Validation={Nval}, N_Test={Nte} (Total={Np})")

# Define the training set (40%) - used for normalization stats
X_tr_df=Xsh[0:Ntr]


#%% evaluate mean and st.dev. for the (true) training data only
# (Page 5, point 4)
# These stats (mm, ss) will be used to normalize ALL data (Train, Val, Test)
print("---------------------------------------")
print("---------------------------------------")
print("Calculating mean/std from TRUE training set (first 40%)...")
mm=X_tr_df.mean()
ss=X_tr_df.std()
# store mean and std for the regressand (total_UPDRS) to denormalize predictions later
my=mm['total_UPDRS']
sy=ss['total_UPDRS']


#%% Generate the normalized/scaled training and test datasets, remove unwanted regressors
# Normalize full shuffled dataset using training statistics (avoid data leakage)
Xsh_norm_df=(Xsh-mm)/ss                                                # normalized data

# keep the normalized target separately
ysh_norm_df=Xsh_norm_df['total_UPDRS']                                    # regressand only

# Define the columns to drop (all columns NOT in the 17 regressors list)
columns_to_drop = [
    'total_UPDRS',  # The regressand (target)
    'subject#',     # Not a feature
    'test_time',    # Not a feature
    'Jitter:DDP',   # Not in the PDF list
    'Shimmer:DDA'   # Not in the PDF list
]

# drop the target and non-regressor columns before fitting
Xsh_norm_df=Xsh_norm_df.drop(columns_to_drop, axis=1)  # regressors only


#%% Define the list of regressors
regressors=list(Xsh_norm_df.columns)
Nf = len(regressors)                                                # number of regressors
print("---------------------------------------")
print("---------------------------------------")
print(f"Feature selection complete. Using {Nf} regressors.")
#print(regressors)

# convert dataframes/series to numpy arrays for linear algebra operations
Xsh_norm=Xsh_norm_df.values                                            # from dataframe to Ndarray
ysh_norm=ysh_norm_df.values                                            # from dataframe to Ndarray

# split normalized arrays into training, validation, and test sets [cite: 49, 54]
X_tr_norm=Xsh_norm[0:Ntr]                                           # regressors for training phase
X_val_norm=Xsh_norm[Ntr:(Ntr+Nval)]
X_te_norm=Xsh_norm[(Ntr+Nval):]                                      # regressors for test phase
y_tr_norm=ysh_norm[0:Ntr]
y_val_norm=ysh_norm[Ntr:(Ntr+Nval)]                                  # regressand for validation phase
y_te_norm=ysh_norm[(Ntr+Nval):]                                      # regressand for test phase

print("---------------------------------------")
print("---------------------------------------")
print(f"SHAPE TRAIN SET:{X_tr_norm.shape},\nSHAPE VALIDATION SET:{X_val_norm.shape}, \nSHAPE TEST SET:{X_te_norm.shape}")
print(f"SHAPE TRAIN TARGET:{y_tr_norm.shape},\nSHAPE VALIDATION TARGET:{y_val_norm.shape}, \nSHAPE TEST TARGET:{y_te_norm.shape}")



#FINISH PREPARATION
```

```python
#STARTING K-EVALUATION



print("-------------------------------------------------------------------------")
print("-------------------------------------------------------------------------")
print("-------------------üß†‚Äã OPTIMIZATION OF K üß†‚Äã-----------------------------")
print("-------find the best k(hyperparameter over the validation set-----------")
print("-------------------------------------------------------------------------")


#%% --- OPTIMIZATION OF K ---
# Find K_opt by looping K and finding the minimum MSE on the VALIDATION set.
print("\n--- Starting KNN Regression (Local Ridge) K-Sweep ---")

# Get the true target values for the validation set
y_true_val_array = y_val_norm
I = np.eye(Nf) # Identity matrix of shape (Nf, Nf)

# Lists to store the results of the K-sweep
k_values = []
mse_per_k = []

# --- EXPLANATION for 'range(Nf + 1, 201)' ---
# We are looping from K = 18 to 200.
# Why start at K=18?
# The number of features (Nf) is 17.
# In step 1.3, we must compute: w_hat = inv(A.T @ A + eps*I) @ A.T @ y
# The matrix to invert, (A.T @ A), has shape (Nf, Nf), which is (17, 17).
# The matrix 'A_local' (A) has shape (K, Nf), which is (K, 17).
#
# If K < Nf (e.g., K=10, Nf=17), 'A_local' is a "wide" matrix.
# The matrix (A.T @ A) will be RANK-DEFICIENT. Its rank can be at most K.
# A rank-deficient matrix is SINGULAR (determinant is 0) and cannot be inverted.
# This will cause a 'LinAlgError: Singular matrix'.
#
# The 'epsilon * I' term (from Ridge Regression) adds a small value to the
# diagonal, which *should* make the matrix invertible (full-rank).
# However, if K is small and epsilon is tiny, the matrix is still "ill-conditioned"
# and can be numerically unstable, leading to errors or nonsensical results.
#
# By starting at K = Nf + 1 (K=18), we ensure K > Nf.
# This guarantees that 'A_local' is a "tall" matrix (or square),
# and (A.T @ A) is very likely to be full-rank and easily invertible.
# We must use at least as many neighbors (K) as we have features (Nf)
# to get a stable solution for the local linear model.
#
# We start the loop at Nf + 1 = 18.
# ----------------------------------------------------
for k in range(Nf + 1, 201):

    k_values.append(k)
    total_squared_error_for_this_k = 0.0

    # Iterate through each sample 'x' in the validation dataset
    for i in range(len(X_val_norm)):
        x_val = X_val_norm[i]     # Current validation sample (shape (Nf,))
        y_true = y_true_val_array[i] # True UPDRS value for this sample

        # 1.1 Find the square distance between x_val and all training points
        sq_distances = np.sum((X_tr_norm - x_val)**2, axis=1) # Shape (Ntr,)

        # 1.2 Select the K points from training with the smallest distance
        k_nearest_indices = np.argsort(sq_distances)[:k]

        # 1.3 Generate local matrix A (K, Nf) and local vector y (K, 1)
        A_local = X_tr_norm[k_nearest_indices]
        y_local = y_tr_norm[k_nearest_indices].reshape(-1, 1)

        # 1.3 Generate w_hat using Ridge Regression formula
        A_T_A = A_local.T @ A_local     # Shape (Nf, Nf)
        A_T_y = A_local.T @ y_local     # Shape (Nf, 1)

        # Solve for w_hat
        w_hat = np.linalg.inv(A_T_A + epsilon * I) @ A_T_y # Shape (Nf, 1)

        # 1.4 Evaluate the estimated UPDRS for the validation point x_val
        y_pred = x_val @ w_hat  # (Nf,) @ (Nf, 1) -> (1,)
        y_pred_scalar = y_pred[0] # Get scalar value

        # 1.4 Calculate and accumulate the estimation error
        error = y_true - y_pred_scalar
        total_squared_error_for_this_k += error**2

    # 2. Evaluate the Mean Square Error for this K on the validation dataset
    mse_val_knn = total_squared_error_for_this_k / len(X_val_norm)
    mse_per_k.append(mse_val_knn)

# --- End of K-loop ---

# 5. Find K_opt and Plot
best_mse = np.min(mse_per_k)
best_k_index = np.argmin(mse_per_k)
best_k = k_values[best_k_index] # This is K_opt


print("--- KNN Regression K-Sweep Complete ---")
print("---------------------------------------")
print("---------------------------------------")
print(f"Optimal K (K_opt): {best_k}")
print("---------------------------------------")
print("---------------------------------------")
print(f"Validation Mean Square Error (MSE) at K_opt: {best_mse:.6f}")

# Plot the MSE vs. K
plt.figure(figsize=(7,5))
plt.plot(k_values, mse_per_k)
plt.xlabel('K (Number of Neighbors)')
plt.ylabel('Mean Square Error (MSE) on Validation Set')
plt.title('KNN Local Regression: MSE vs. K')
plt.grid(True)
plt.axvline(x=best_k, color='red', linestyle='--', label=f'Best K (K_opt) = {best_k}')
plt.legend()
#plt.savefig('knn_mse_vs_k.png') # Save the plot



#FINISH K-EVALUATION
```

```python
#STARTING TEST-PHASE



print("-------------------------------------------------------------------------")
print("-------------------------------------------------------------------------")
print("----------------------------üß†‚Äã TEST PHASE üß†‚Äã-----------------------------")
print(f"-------------Using K_opt = {best_k} to predict on test set...-----------")
print("-------------------------------------------------------------------------")

#%% --- TEST PHASE ---
# (Page 8)
# Now we use K_opt (best_k) to evaluate performance on the TEST SET.
print(" We will compare:")
print(f"# 1. KNN-LLS (K=K_opt) trained on 40% (X_tr_norm)")
print(f"# 2. Standard LLS trained on 60% X_tr_norm + X_val_norm)   (LAB 1)")


# 1. Get predictions for KNN-LLS (Page 8, point 1)
# Trained on X_tr_norm, predicting X_te_norm

# We inline the prediction loop here, matching the logic used in the validation phase
# This avoids the external helper function.

I_mat = np.eye(Nf)
y_predictions_knn_test = [] # Lista per raccogliere le predizioni sul test set

# Iteriamo direttamente su ogni campione (vettore feature) del TEST set
for x_pred in X_te_norm:
    # 1.1 Find distances (compute distance from x_pred to all points in X_tr_norm)
    sq_distances = np.sum((X_tr_norm - x_pred)**2, axis=1)

    # 1.2 Get K-nearest (using best_k)
    k_nearest_indices = np.argsort(sq_distances)[:best_k]

    # 1.3 Generate local matrices (using the neighbors from the training set)
    A_local = X_tr_norm[k_nearest_indices]
    y_local = y_tr_norm[k_nearest_indices].reshape(-1, 1)

    # 1.3 Solve for w_hat (local weights)
    A_T_A = A_local.T @ A_local
    A_T_y = A_local.T @ y_local
    w_hat = np.linalg.inv(A_T_A + epsilon * I_mat) @ A_T_y

    # 1.4 Predict
    y_pred = x_pred @ w_hat
    y_predictions_knn_test.append(y_pred[0])

# Convert predictions list to numpy array
y_hat_te_norm_KNN = np.array(y_predictions_knn_test)
print("---------------------------------------")
print("---------------------------------------")
print("# 1.")
print("Test predictions for KNN-LLS computed.")


# 2. Get predictions for Standard LLS (Page 8, point 3)
# Trained on 60% of data (Train + Val)
print("---------------------------------------")
print("---------------------------------------")
print("# 2.")
print("Training standard LLS on 60% of data (Train+Val)...")
X_tr_full = np.vstack((X_tr_norm, X_val_norm)) # 60% of features
y_tr_full = np.hstack((y_tr_norm, y_val_norm)) # 60% of targets


print(f"Full training set shape: {X_tr_full.shape} (no train/validation division)")

w_hat_LLS_full = np.linalg.inv(X_tr_full.T @ X_tr_full) @ (X_tr_full.T @ y_tr_full)
y_hat_te_norm_LLS = X_te_norm @ w_hat_LLS_full # Predict on test set

print("Test predictions for Standard LLS computed.")


#%% De-normalize all test data for comparison
# (Page 8, point 2 & 4)

# True test values
y_te = y_te_norm * sy + my
# LLS predictions
y_hat_te_LLS = y_hat_te_norm_LLS * sy + my
# KNN-LLS predictions
y_hat_te_KNN = y_hat_te_norm_KNN * sy + my









#%% Statistics of the errors (KNN-LLS)
print("---------------------------------------")
print("---------------------------------------")
print("# 1.")
print(f" Statistics for KNN-LLS (K_opt = {best_k})")
E_te_KNN = (y_te - y_hat_te_KNN) # Errors on the test dataset

E_te_max_KNN = E_te_KNN.max()
E_te_min_KNN = E_te_KNN.min()
E_te_mu_KNN = E_te_KNN.mean()
E_te_sig_KNN = E_te_KNN.std()
E_te_MSE_KNN = np.mean(E_te_KNN**2)
R2_te_KNN = 1 - E_te_MSE_KNN / (np.var(y_te))
c_te_KNN = np.mean((y_te - y_te.mean()) * (y_hat_te_KNN - y_hat_te_KNN.mean())) / (y_te.std() * y_hat_te_KNN.std())

cols = ['min', 'max', 'mean', 'std', 'MSE', 'R^2', 'corr_coeff']
rows = ['Test (KNN-LLS)']
p_KNN = np.array([[E_te_min_KNN, E_te_max_KNN, E_te_mu_KNN, E_te_sig_KNN, E_te_MSE_KNN, R2_te_KNN, c_te_KNN]])
results_KNN = pd.DataFrame(p_KNN, columns=cols, index=rows)
print(results_KNN)







#%% Statistics of the errors (LLS)
print("---------------------------------------")
print("---------------------------------------")
print("# 2.")
print(" Statistics for Standard LLS (trained on 60%)")
E_te_LLS = (y_te - y_hat_te_LLS) # Errors on the test dataset

E_te_max_LLS = E_te_LLS.max()
E_te_min_LLS = E_te_LLS.min()
E_te_mu_LLS = E_te_LLS.mean()
E_te_sig_LLS = E_te_LLS.std()
E_te_MSE_LLS = np.mean(E_te_LLS**2)
R2_te_LLS = 1 - E_te_MSE_LLS / (np.var(y_te))
c_te_LLS = np.mean((y_te - y_te.mean()) * (y_hat_te_LLS - y_hat_te_LLS.mean())) / (y_te.std() * y_hat_te_LLS.std())

cols = ['min', 'max', 'mean', 'std', 'MSE', 'R^2', 'corr_coeff']
rows = ['Test (LLS)']
p_LLS = np.array([[E_te_min_LLS, E_te_max_LLS, E_te_mu_LLS, E_te_sig_LLS, E_te_MSE_LLS, R2_te_LLS, c_te_LLS]])
results_LLS = pd.DataFrame(p_LLS, columns=cols, index=rows)

print(results_LLS)




print("---------------------------------------")
print("---------------------------------------")
print("\n--- Comparison ---")
if E_te_MSE_KNN < E_te_MSE_LLS:

    print(f"Technique is the best: KNN-LLS (MSE: {E_te_MSE_KNN:.4f} vs {E_te_MSE_LLS:.4f})")
else:

    print(f"Technique is the best: Standard LLS (MSE: {E_te_MSE_LLS:.4f} vs {E_te_MSE_KNN:.4f})")

print(f"KNN-LLS R^2: {R2_te_KNN:.4f} vs Standard LLS R^2: {R2_te_LLS:.4f}")






#FINISH TEST PHASE
```

```python
#START PLOTTING

#%% plot the error histograms (Comparison)
# This script plots a comparison histogram to visualize and compare the
# test error distributions of the Standard LLS and KNN-LLS models.

# Find the global min/max error across BOTH models for a shared axis
M = np.max([np.max(E_te_LLS), np.max(E_te_KNN)])
m = np.min([np.min(E_te_LLS), np.min(E_te_KNN)])

# Create 50 common bins to ensure a fair comparison
common_bins = np.arange(m, M, (M-m)/50)

# Prepare data and labels for the plot
e_comp = [E_te_LLS, E_te_KNN]
labels_comp = ['Standard LLS (60% train)', f'KNN-LLS (K_opt={best_k})']

# Create the figure
plt.figure(figsize=(7,5))

# Plot both histograms, using density=True to normalize them
plt.hist(e_comp, bins=common_bins, density=True, histtype='bar', label=labels_comp)

# Add labels, title, and legend
plt.xlabel(r'$e=y-\^y$ (Test Error)')
plt.ylabel(r'$P(e$ in bin$)$')
plt.legend()
plt.grid()
plt.title('Test Error Histograms Comparison')
plt.tight_layout() # Adjust layout to prevent overlap
#plt.savefig('./TEST-hist-comparison.png')


#%% plot the regression line for Standard LLS
plt.figure(figsize=(5,5))
plt.plot(y_te, y_hat_te_LLS, '.', label=f'Standard LLS (R^2={R2_te_LLS:.3f})', alpha=0.6, color='blue')
v = plt.axis() # Get current axis limits
plt.plot([v[0], v[1]], [v[0], v[1]], 'r', linewidth=2, label='Ideal (y_hat = y)')
plt.legend()
plt.xlabel(r'$y_{true}$')
plt.axis('square') # Ensure x and y axes have the same scale
plt.ylabel(r'$\^y_{predicted}$')
plt.grid()
plt.title('Test Set: Standard LLS (60% Train)')
plt.tight_layout()
#plt.savefig('./TEST-yhat_vs_y_LLS.png')

#%% plot the regression line for KNN-LLS
plt.figure(figsize=(5,5))
plt.plot(y_te, y_hat_te_KNN, '.', label=f'KNN-LLS K={best_k} (R^2={R2_te_KNN:.3f})', alpha=0.6, color='green')
v = plt.axis() # Get current axis limits
plt.plot([v[0], v[1]], [v[0], v[1]], 'r', linewidth=2, label='Ideal (y_hat = y)')
plt.legend()
plt.xlabel(r'$y_{true}$')
plt.axis('square') # Ensure x and y axes have the same scale
plt.ylabel(r'$\^y_{predicted}$')
plt.grid()
plt.title(f'Test Set: KNN-LLS (K_opt={best_k})')
plt.tight_layout()
#plt.savefig('./TEST-yhat_vs_y_KNN.png')

# Show all generated plots
plt.show()




```

```python
"""




Ecco i tre motivi principali per cui la **Ridge Regression**
(il termine `+ epsilon * I`) √® **obbligatoria** nel KNN-LLS
ma opzionale (anche se consigliata) nel LLS standard:


### 1. Il Problema della Dimensione (Rank Deficiency)

* **LLS Standard (Globale):** Hai circa **2000+ righe** (pazienti) per
risolvere **17 colonne** (features).
La matrice  √® costruita su una massa enorme di dati. √à statisticamente molto
improbabile che 2000 punti siano perfettamente allineati in modo da rendere la
matrice non invertibile (singolare). Il sistema √® *sovradeterminato*.
* **KNN-LLS (Locale):** Qui usi solo ** righe** per risolvere **17 colonne**.
Il codice permette di testare  a partire da 18. Se  √® basso (es. 18, 20, 25),
 stai cercando di risolvere un sistema con un numero di equazioni (righe) quasi
 uguale al numero di incognite (colonne). Basta una minima ridondanza nei dati
 per rendere la matrice matematicamente impossibile da invertire (determinante = 0).

### 2. Il Paradosso dei "Vicini" (Multicollinearit√† Estrema)

Nel metodo KNN, selezioni intenzionalmente punti che sono **molto simili tra
loro** (sono "vicini").

* Essendo simili, le loro righe nella matrice sono quasi identiche.
* In algebra lineare, righe identiche o quasi identiche non portano nuove
informazioni.
* Questo rende la matrice  **mal condizionata** (ill-conditioned).
Senza la Ridge (`epsilon`), il computer proverebbe a dividere per zero o
genererebbe pesi  giganteschi e privi di senso.

### 3. Stabilit√† dell'Algoritmo nel Loop

Nel codice c'√® un ciclo che testa vari .
Senza il termine `epsilon * I`, se il ciclo provasse un  molto piccolo o
capitasse in una zona dello spazio dove i pazienti sono tutti uguali, il
codice andrebbe in crash con l'errore:
`LinAlgError: Singular matrix`

**In sintesi:**
Nel LLS globale la matrice √® solida grazie alla mole di dati. Nel KNN locale
la matrice √® fragile e instabile perch√© basata su pochi punti molto simili; l
a Ridge (`epsilon`) serve come "colla" matematica per impedire che il calcolo esploda.
"""
```
